# PREPROCESSING

```python
# split data into train and test sets
X = df.drop(columns = ['Heart Attack Risk'])
y = df['Heart Attack Risk']
# train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.2, stratify=y)
```

```python
# check the class balance
y_train.value_counts(normalize=True)
```

```python
y_test.value_counts(normalize=True)
```

```python
# create selectors for preprocessing
cat_selector = make_column_selector(dtype_include = 'object')
num_selector = make_column_selector(dtype_include = 'number')
# Create transformers
scaler = StandardScaler()
ohe = OneHotEncoder(sparse=False, handle_unknown='ignore')
# pipelines
num_pipe = make_pipeline(scaler)
cat_pipe = make_pipeline(ohe)
# tuples
num_tuple = (num_pipe, num_selector)
cat_tuple = (cat_pipe, cat_selector)
#column transformer
column_transformer =make_column_transformer(num_tuple, cat_tuple, remainder = 'passthrough')
```

```python
# ColumnTransformer
preprocessor = make_column_transformer(num_tuple, cat_tuple)
preprocessor
```

```python
# fit on train
preprocessor.fit(X_train)
```

# MODELLING

```python
# Creating a pipeline for KNN
pipeline_knn = make_pipeline(preprocessor, KNeighborsClassifier())

# Fitting the KNN pipeline
pipeline_knn.fit(X_train, y_train)

# Predicting on the training and test sets
y_pred_knn_train = pipeline_knn.predict(X_train)
y_pred_knn = pipeline_knn.predict(X_test)

# Evaluation metrics for KNN
print("\nK-Nearest Neighbors - Default Hyperparameters:")
print("Accuracy Train:", accuracy_score(y_train, y_pred_knn_train))
print("Accuracy Test:", accuracy_score(y_test, y_pred_knn))
print("Classification Report:\n", classification_report(y_test, y_pred_knn))
```

```python
# make a RandomForestClassifier instance
rf = RandomForestClassifier()
# create pipeline for the model
pipeline_rf = make_pipeline(preprocessor, rf)
pipeline_rf.fit(X_train, y_train)
```

```python
# evaluation of the rf model
y_pred = pipeline_rf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
# Make a pipeline for log reg
pipeline_logreg = make_pipeline(preprocessor,
                                LogisticRegression(max_iter=200, random_state=42))

# fit log reg pipeline
pipeline_logreg.fit(X_train, y_train)

# Predicting on the test set
y_pred_logreg_train= pipeline_logreg.predict(X_train)
y_pred_logreg = pipeline_logreg.predict(X_test)
# evaluation metrics
print("\nLogistic Regression - Default Hyperparameters:")
print("Accuracy train:", accuracy_score(y_train, y_pred_logreg_train))
print("Accuracy test:", accuracy_score(y_test, y_pred_logreg))
print("Classification Report:\n", classification_report(y_test, y_pred_logreg))
```

# TUNING THE MODELS

```python
# Tuning KNN Model
param_grid_knn = {
    'kneighborsclassifier__n_neighbors': [3, 5, 7, 9],
}

grid_search_knn = GridSearchCV(pipeline_knn, param_grid_knn, cv=5)
grid_search_knn.fit(X_train, y_train)

# Best hyperparameters for KNN
print("Best Hyperparameters for K-Nearest Neighbors:", grid_search_knn.best_params_)

# Prediction on the tuned KNN model
y_pred_knn_tune_train = grid_search_knn.predict(X_train)
y_pred_knn_tuned = grid_search_knn.predict(X_test)

# Evaluation metrics for the tuned KNN model
print("\nK-Nearest Neighbors - Best Hyperparameters:")
print("Accuracy Train:", accuracy_score(y_train, y_pred_knn_tune_train))
print("Accuracy Test:", accuracy_score(y_test, y_pred_knn_tuned))
print("Classification Report:\n", classification_report(y_test, y_pred_knn_tuned))
```

```python
# tuning random forest classifier
param_grid_rf = {
    'randomforestclassifier__n_estimators': [10, 50, 100, 200],
    'randomforestclassifier__max_depth': [None, 10, 20, 30, 40, 50],
}
# tuning random forest classifier
param_grid_rf = {
    'randomforestclassifier__n_estimators': [10, 50, 100, 200],
    'randomforestclassifier__max_depth': [None, 10, 20, 30, 40, 50],
}

grid_search_rf = GridSearchCV(pipeline_rf, param_grid_rf, cv=5)
grid_search_rf.fit(X_train, y_train)
y_pred_rf_tuned = grid_search_rf.predict(X_test)
y_pred_rf_tuned_train = grid_search_rf.predict(X_train)
# evaluation metrics
print("\nRandom Forest - Best Hyperparameters:", grid_search_rf.best_params_)
print("Accuracy train:", accuracy_score(y_train, y_pred_rf_tuned_train))
print("Accuracy test:", accuracy_score(y_test, y_pred_rf_tuned))
print("Classification Report:\n", classification_report(y_test, y_pred_rf_tuned))

```

```python
# Tuning log reg model
param_grid = {
    'logisticregression__C': [0.001, 0.01, 0.1, 1, 10, 100, 1000],
    'logisticregression__penalty': ['l1', 'l2'],
    'logisticregression__solver': ['liblinear']
}

grid_search_logreg = GridSearchCV(pipeline_logreg, param_grid, cv=5)
grid_search_logreg.fit(X_train, y_train)

# Best hyperparameters
print("Best Hyperparameters for Logistic Regression:", grid_search_logreg.best_params_)

# prediction on the tuned model for log reg
y_pred_logreg_tune_train = grid_search_logreg.predict(X_train)
y_pred_logreg_tuned = grid_search_logreg.predict(X_test)

# eval metrics
print("\nLogistic Regression - Best Hyperparameters:")
print("Accuracy Train:", accuracy_score(y_train, y_pred_logreg_tune_train))
print("Accuracy test:", accuracy_score(y_test, y_pred_logreg_tuned))
print("Classification Report:\n", classification_report(y_test, y_pred_logreg_tuned))
```
