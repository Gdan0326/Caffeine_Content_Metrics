```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load data
file ='/content/drive/MyDrive/Colab Notebooks/Data/Coffee Caffeine Content.xlsx'
df = pd.read_excel(file)
df.head(20)
```

```python
# Inspect for duplicated and missing data
df.isnull().sum()
df.duplicated()
```

```python
# Check data types
df.dtypes
```

```python
# Check the summary statistics of caffeine content and drink size
df.describe()
```

```python
# Visualize average coffee content accross coffee retailers using bar plot
Chain = df['Chain']
Caffeine = df['Caffeine (mg)']
Drink_size = df['Drink size (ml)']
sns.barplot (x = Chain, y = Caffeine, data = df);
plt.title('Average Caffeine Content by Product across Coffee Retailers', fontsize = 18);
plt.xlabel ('Coffee Chain', fontsize = 16);
plt.ylabel ('Average Caffeine Content (mg)', fontsize = 16);
```

```python
# from the above graph it can be inferred that Costa and Pret reflected the highest average caffeine content of around 210 and 205, respectively  for its products
# Gregg's coffee chain come in third garnering an average caffeine content of around 175 mg
# Caffe Nero and Starbucks reflected low average caffeine content of around 80 mg and 70 mg, respectively
```

```python
# scatterplot showing the relationship of caffeine content and drink size
plt.figure(figsize =(8, 5))
sns.scatterplot(x='Drink size (ml)', y=Caffeine, hue='Chain', data=df, palette='coolwarm', s=100);
plt.title('Caffeine Content vs Drink Size', fontsize=16)
plt.xlabel('Drink Size (ml)', fontsize=14)
plt.ylabel('Caffeine Content (mg)', fontsize=14)
plt.legend(title='Chain', loc=(1.02, 0))
```

```python
# The scatterplot shows that there is no clear linear relationship between the drink size and its caffeine content
# Small drink sizes around 30 ml displays varying caffeine content from 20 mg to around 180 mg
#  from the graph above it can be inferred  that there is variation in both drink size and caffeine content across products from different coffee retailers
# There is a wide range of option to cater different consumer preference like comsumers opting for smaller drink size but high caffeine content and vice versa
```
